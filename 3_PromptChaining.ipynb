Prompt chaining -  series me multiple calls karte  hai hum apne task ko small small steps me todkr hum esries me calls kete hai

```python
import os
from dotenv import load_dotenv
from langgraph.graph import StateGraph, START, END
from langchain_groq import ChatGroq
load_dotenv()

llm = ChatGroq(model='llama-3.1-8b-instant')
```

```python
from typing import TypedDict
class BlogState(TypedDict):
    title: str
    content: str
    conclusion: str
    outline: str
```

```python
def Outli(state: BlogState) -> BlogState:
    title = state['title']

    outline = llm.invoke(f"generate detailed outline for the blog on title {title}").content

    state['outline'] = outline
    return state
```

```python
def cont(state: BlogState)-> BlogState:
    title= state['title']
    outline = state['outline']

    content = llm.invoke(f"Write a main content on topic {title} and outline {outline}").content

    state['content'] = content
    return state
```

```python
def conclus(state: BlogState) -> BlogState:
    title= state['title']
    outline = state['outline']
    content = state['content']

    conclusion = llm.invoke(f"Generate a conclusion on topic {title} , outline {outline} and content {content}").content
    state['conclusion']= conclusion
    return state
```

```python
graph = StateGraph(BlogState)

graph.add_node("outline_node", Outli)
graph.add_node("content_node", cont)
graph.add_node("conclusion_node", conclus)

graph.add_edge(START, "outline_node")
graph.add_edge("outline_node", "content_node")
graph.add_edge("content_node", "conclusion_node")
graph.add_edge("conclusion_node", END)

workflow = graph.compile()
initial_state = {"title": "take ai as a competitor"}
final_state = workflow.invoke(initial_state)
print(final_state)
```
