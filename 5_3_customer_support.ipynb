```python
from langchain_groq  import ChatGroq
from langgraph.graph import StateGraph, START, END
from typing import TypedDict, Literal
from pydantic import BaseModel, Field
from dotenv import load_dotenv
load_dotenv()
llm  = ChatGroq(model="llama-3.1-8b-instant")
```

```python
class SentimentSchema(BaseModel):
    sentiment: Literal["positive", "negative"] = Field(description="Sentiment of the customer")

structured_llm = llm.with_structured_output(SentimentSchema)
```

```python
def get_sentiment(state: SentimentSchema):
    prompt = f"Get the sentiment of the customer by this question,{state["query"]}, "
    sentiment = structured_llm.invoke(prompt)
    return {"sentiment": sentiment} 
```

```python
structured_llm = llm.with_structured_output(SentimentSchema)
prompt="the software is good"
print(structured_llm.invoke(prompt))
```

```python
class customer_service(BaseModel):
    query: str = Field(description= "the query of customer")
    response: str = Field(description=" Respponse from company")
    tone: str = Field(description=" tone of customer")
    sentiment: Literal["positive", "negative"]
    diagnosis: str
```

```python
def conditional_routing(state: SentimentSchema):
    sentiment = state["sentiment"]
    if sentiment== 'positive'
        return "positive_sen"
    else:
        return "negative_sen"
```
